```python
import matplotlib.pyplot as plt
%matplotlib inline

levs = d.coords['lev']
times = d.coords['time']

for t in times:
    fig, axes = plt.subplots(ncols=3, figsize=(14, 4))
    s = {'time': t, 'lev': levs[0]}
    ds = d.sel(s)
    clct, u_10, v_10  = ds['clct'], ds['10u'][0], ds['10v'][0]
    for i, f in enumerate([clct, u_10, v_10]):
        f.plot.contourf('lon', 'lat', ax=axes[i])
        axes[i].set_title("{} at {}".format(f.name, t.values))
    plt.show()
```
